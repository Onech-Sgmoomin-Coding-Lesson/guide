```python
import tkinter as tk
from tkinter import Label, Button
import requests
import json

def get_weather():
    city = city_entry.get()
    apikey = "a955cb096d0c9850486d8825884d4f72"
    lang = "kr"
    
    api = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={apikey}&lang={lang}&units=metric"
    
    result = requests.get(api)
    data = json.loads(result.text)

    weather_info.config(text=f"{data['name']}의 날씨\n"
                            f"날씨: {data['weather'][0]['description']}\n"
                            f"현재 온도: {data['main']['temp']}°C\n"
                            f"체감 온도: {data['main']['feels_like']}°C\n"
                            f"최저 기온: {data['main']['temp_min']}°C\n"
                            f"최고 기온: {data['main']['temp_max']}°C\n"
                            f"습도: {data['main']['humidity']}%\n"
                            f"기압: {data['main']['pressure']} hPa\n"
                            f"풍향: {data['wind']['deg']}°\n"
                            f"풍속: {data['wind']['speed']} m/s")

# 메인 창 생성
window = tk.Tk()
window.title("날씨 정보 조회")

# 위젯 생성 및 구성
city_label = Label(window, text="도시:")
city_label.pack(pady=5)

city_entry = tk.Entry(window)
city_entry.pack(pady=5)

get_weather_button = Button(window, text="날씨 조회", command=get_weather)
get_weather_button.pack(pady=10)

weather_info = Label(window, text="")
weather_info.pack(pady=10)

# GUI 이벤트 루프 시작
window.mainloop()
```
